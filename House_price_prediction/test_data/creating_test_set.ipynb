```python
def shuffle_and_split(data,test_ratio):
    np.random.seed(42) #set the seed for reproducibility
    shuffled_indices = np.random.permutation(len(data)) #returns shuffle indices
    test_set_size = int(len(data) * test_ratio) #test ration should not into percentage
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=  1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(df,df['income_cat']):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
```

```python
#training data will represent for entire population
import matplotlib.pyplot as plt
strat_train_set["income_cat"].value_counts().plot.bar(rot=0, grid=True)
plt.title("Income Categories Distribution")
plt.xlabel("Income Category")
plt.ylabel("Number of Instances")
```
