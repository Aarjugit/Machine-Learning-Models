```python
import pandas as pd
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
data = pd.read_excel("iris_train.xlsx")
```

```python
data.head()
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
y
```

```python
model=GaussianNB()
```

```python
model.fit(x.values,y.values)
```

```python
datatest=pd.read_excel("iris_test.xlsx")
```

```python
x_test=datatest.iloc[:,:-1]
```

```python
predictions = model.predict(x_test.values)
```

```python
predictions
```

```python
#Now we need to check test fro y 
y_test=datatest.iloc[:,-1]
```

```python
#Will make array with values
Actual=y_test.values
```

```python
#Now I need to compare prediction and real labeled data and I have 2 arrays.
count = 0
for i in range(0,len(Actual)):
    if Actual[i] == predictions[i]:
        count= count+1
#we will find Accuracy percentage
print((count*100)/len(Actual))
```
